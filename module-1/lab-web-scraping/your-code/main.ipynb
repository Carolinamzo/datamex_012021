#### Display all the image links from Walt Disney wikipedia page

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
#your code
html = requests.get(url).content
soup = bs(html, 'lxml')
````

````python
link1 = []
for i in soup.findAll('img', {'src':re.compile('.JPG')}):
    link1.append(i.get('src'))
    
link1
````

````python
link2 = []
for j in soup.findAll('img', {'src':re.compile('.jpg')}):
    link2.append(j.get('src'))
    
link2
````

````python
link3 = []
link4 = []
for a in link1:
    link3.append('https://' + a)

for b in link2:
    link4.append('https://' + b)
    
link3+link4
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
html = requests.get(url).content
soup = bs(html, 'lxml')
````

````python
info_date = soup.findAll('td', {'class':'tabev6'})[:20]
info_lat = soup.findAll('td', {'class':'tabev1'})[:40]
info_reg = soup.findAll('td', {'class':'tb_region'})[:20]
````

````python
lat_long = [i.text.strip().split('\n') for i in info_lat]

lat = lat_long[0::2]
long = lat_long[1::2]

print(lat, long)
````

````python
region = [j.get_text().strip().split('\n') for j in info_reg]
region
````

````python
datetime = [k.text.strip().split('\n') for k in info_date]

datetime
````

````python
date = []
time = []

for a in range(len(datetime)):
    path_date = '[0-9]+-[0-9]+-[0-9]+'
    path_time = '[0-9]+:[0-9]+:[0-9]+'
    date.extend([re.findall(path_date, a) for a in datetime[a]])
    time.extend([re.findall(path_time, a) for a in datetime[a]])
    
columns = [date, time, lat, long, region]
col = [[a for [a] in b] for b in columns]
````

````python
earthquake = pd.DataFrame(list(zip(col[0], col[1], col[2], col[3], col[4])),
                         columns = ['date','time','lat','long', 'region'])
````

````python
earthquake
````
